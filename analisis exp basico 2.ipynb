```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
adult = fetch_ucirepo(id=2)

# data (as pandas dataframes)
X = adult.data.features
y = adult.data.targets

# metadata
print(adult.metadata)

# variable information
print(adult.variables)
```

```python
import pandas as pd
data = pd.concat([X,y],axis=1)
data
```

```python
import numpy as np
data['income'] = np.where(data['income'] == '<=50K.', '<=50K', data['income'])
data['income'] = np.where(data['income'] == '>50K.', '>50K', data['income'])
data['income'].value_counts(dropna=False)
```

```python
# Agrupando las edades
data['age_agrup'] = '50 ++'
data['age_agrup'] = np.where(data['age'] <= 50, '30 50', data['age_agrup'])
data['age_agrup'] = np.where(data['age'] <= 30, '-- 30', data['age_agrup'])
data['age_agrup'].value_counts()
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Histogram(x=data['age'][(data['income']=='<=50K')], name='<=50K', opacity=0.75))
fig.add_trace(go.Histogram(x=data['age'][(data['income']=='>50K')], name='>50K', opacity=0.75))
fig.update_layout(barmode='overlay')

fig.show()
```

```python
data['workclass'] = data['workclass'].fillna('Vacios')
data['workclass'].value_counts(dropna=False)
```

```python
pd.crosstab(data['workclass'], data['income'])
```

```python
tab = pd.crosstab(data['workclass'], data['income'])
tab = tab.reset_index()
tab['Total'] = tab['<=50K'] + tab['>50K']
tab['<=50K'] = tab['<=50K'] / tab['Total']
tab['>50K'] = tab['>50K'] / tab['Total']
tab
```
