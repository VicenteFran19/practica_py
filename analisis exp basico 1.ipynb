```python
# Analisis de RegresiÃ³n
```

```python
import statsmodels.api as sm
X = data.drop(columns=['price'])
Y = data['price']
```

```python
# agregando la constante para tener el B0
X = sm.add_constant(X)
```

```python
# Ejecutar modelo de regresion
model = sm.OLS(Y, X).fit()
print(model.summary())
```

```python
# Modelo v2 (eliminando 'mileage' por correlaciones)

X = data.drop(columns=['price','mileage'])
X = sm.add_constant(X)
Y = data['price']

model = sm.OLS(Y, X).fit()
print(model.summary())
```
